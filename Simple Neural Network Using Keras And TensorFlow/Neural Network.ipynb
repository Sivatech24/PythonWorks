```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
```

# Load the MNIST dataset

```python
mnist = keras.datasets.mnist
(x_train, y_train), (x_test, y_test) = mnist.load_data()
```

# Normalize the input data

```python
x_train, x_test = x_train / 255.0, x_test / 255.0
```

# Define the neural network model

```python
model = keras.Sequential([
    layers.Flatten(input_shape=(28, 28)),  # Flatten the 28x28 images into a 1D array
    layers.Dense(256, activation='relu'),  # First hidden layer with 256 neurons
    layers.Dense(128, activation='relu'),  # Second hidden layer with 128 neurons
    layers.Dense(64, activation='relu'),   # Third hidden layer with 64 neurons
    layers.Dense(10, activation='softmax')  # Output layer with 10 neurons (one per class)
])
```

# Compile the model

```python
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
```

# Train the model

```python
model.fit(x_train, y_train, epochs=5)
```

# Evaluate the model

```python
test_loss, test_acc = model.evaluate(x_test, y_test, verbose=2)
print(f'\nTest accuracy: {test_acc:.4f}')
```

# Save the trained model

```python
model.save("mnist_model.h5")
print("Model saved as 'mnist_model.h5'")
```
